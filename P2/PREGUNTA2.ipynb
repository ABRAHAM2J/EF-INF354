```python
#importamos la libreria os para acceder a los archivos del sistema
import os
#importamos la libreria matplotlib para poder mostrar las imagenes
import matplotlib.pyplot as plt
#importamos la libreria pillow para poder trabjar con imagenes
from PIL import Image
#importaos la libreria opencv para imagenes
import cv2
import numpy as np
```

```python
#Definimos la ruta donde se encuentran las imagenes
directorio_imagenes = './trainingSample/'
```

```python
#definimos una funcion para obtener las rutas de todas las imagenes
def ruta_imagenes(directorio_imagenes):
    #obtenemos la lista de archivos del directorio donde estan las imagenes
    archivos = os.listdir(directorio_imagenes)
    #obtenemos los archivos jpg
    archivos = [archivo for archivo in archivos if archivo.lower().endswith(('.jpg'))]
    #definimos un array vacio de rutas
    rutas=[]
    #iteramos sobre las archivos
    for i, archivo in enumerate(archivos):
        #obtenemos las rutas completas
        ruta_imagen = os.path.join(directorio_imagenes, archivo)
        #añadimos la ruta al array
        rutas.append(ruta_imagen)
    #devolvemos el array de rutas
    return rutas
rutas_imagenes = ruta_imagenes(directorio_imagenes)
```

```python
#definimos una funcion para mostrar algunas imagenes del dataset
def mostrar_imagenes(rutas_imagenes):
    fig, axes = plt.subplots(1, 5, figsize=(25,5))
    # Mostrar las primeras tres imágenes sin reescalado
    for i in range(5):
        #abrimos la imagen con opencv
        img = cv2.imread(rutas_imagenes[i])
        #mostramos la imagen por cada subplot
        axes[i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        #ponemos titulo a cada subplot
        axes[i].set_title(f'Muestra N°: {i + 1}')
#mostramos 5 imagenes de muestra
mostrar_imagenes(rutas_imagenes)
```

```python
#Mostraremos la matriz de la primera imagen
m1 = cv2.imread(rutas_imagenes[0])
for i in range(len(m1)):
    for j in range(len(m1)):
        print(m1[i][j][0], end=" ")
    print("\n")
m2 = np.array(m1)
print(m2.shape)
```

```python
#definimos una funcion para reescalar las imagenes
def reescalar_imagenes(rutas_imagenes):
    #definimos un array para obtener las imagenes reescaladas
    reescalado = []
    #iteramos sobre el array de rutas
    for i in range(len(rutas_imagenes)):
        #obtenemos las imagenes
        img = cv2.imread(rutas_imagenes[i])
        #reescalamos las imagenes
        res = cv2.resize(img, dsize=(150,150), interpolation=cv2.INTER_CUBIC)
        #añadimos al array
        reescalado.append(res)
    #devolvemos el array
    return reescalado

imagenes_reescaladas = reescalar_imagenes(rutas_imagenes)

#mostraremos las imagenes reescaladas
fig, axes = plt.subplots(1, 5, figsize=(25,5))
# Mostrar las primeras tres imágenes sin reescalado
for i in range(5):
    #mostramos la imagen por cada subplot
    axes[i].imshow(cv2.cvtColor(imagenes_reescaladas[i], cv2.COLOR_BGR2RGB))
    #ponemos titulo a cada subplot
    axes[i].set_title(f'Imagen reescalada N°: {i + 1}')
```

```python
#Mostraremos la matriz de la primera imagen reescalada
imagen_reescalada = imagenes_reescaladas[0]
for i in range(len(imagen_reescalada[0])):
    for j in range(len(imagen_reescalada[0])):
        print(imagen_reescalada[i][j][0], end="")
    print("\n")
m3 = np.array(imagenes_reescaladas[0])
print(m3.shape)
```

```python
#Definiremos una funcion para convertir de matriz a vector
def matriz_vector_reescalado(imagenes_reescaladas):
    #Creamos un array donde estaran las imagenes en forma de vector
    vector_imagenes_filas = []
    #Creamos un array donde estaran las imagenes en forma de vector
    vector_imagenes_columnas = []
    #iteramos sobre el array de iamgenes
    for i in range(len(imagenes_reescaladas)):
        #obtenemos cada una de las imagenes
        imagen = np.array(imagenes_reescaladas[i])
        #añadimos los imagenes convertidas a vector
        vector_imagenes_filas.append(imagen.T.flatten())
        vector_imagenes_columnas.append(imagen.T.flatten(order = 'F'))
    #retornamos los vectores
    return vector_imagenes_filas, vector_imagenes_columnas
```

```python
vector_imagenes_filas, vector_imagenes_columnas = matriz_vector_reescalado(imagenes_reescaladas)
print(len(vector_imagenes_filas[0]))
print(vector_imagenes_filas[0])
print("---------------------------------")
print(vector_imagenes_columnas[0])
```
